```python
# Linear Regression Model

from sklearn.linear_model import LinearRegression

y_train = np.array(df['GDP growth (annual %)']).reshape((-1,1))
x_train = np.array(df['Year']).reshape((-1,1))

x_test = np.array([i for i in range(1971, 2031)]).reshape((-1,1)) 
# years from 1971 - 2030
# 2020-2030 are extra for prediction

lr_model = LinearRegression()
lr_model.fit(x_train, y_train)

print("Coefficient of determination: {}".format(lr_model.score(x_train, y_train)))
print("Slope: {}".format(lr_model.coef_))
print("Intercept: {}".format(lr_model.intercept_))

plt.scatter(x_train, y_train, color = 'red', label = 'Training Set')
plt.plot(x_test, lr_model.predict(x_test), color = 'green', label = 'Prediction')
plt.title('GDP Growth (Annual %)')
plt.xlabel('Year')
plt.ylabel('Percentage Growth')
plt.legend()
plt.grid()
plt.show()
```

```python
# Polynomial Regression Model

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

y_train = np.array(df['GDP growth (annual %)']).reshape((-1,1))
x_train = np.array(df['Year']).reshape((-1,1))

x_test = np.array([i for i in range(1971, 2031)]).reshape((-1,1))
# years from 1971 - 2030
# 2020-2030 are extra for prediction

transformer = PolynomialFeatures(degree = 2, include_bias = False)
transformer.fit(x_train)
x_train_ = transformer.transform(x_train)

# x_train_ = PolynomialFeatures(degree=2, include_bias=False).fit_transform(x_train)
x_test_ = PolynomialFeatures(degree = 2, include_bias = False).fit_transform(x_test)

# print(x_train_)

lr_model = LinearRegression()
lr_model.fit(x_train_, y_train)

print("Coefficient of determination: {}".format(lr_model.score(x_train_, y_train)))
print("Slope: {}".format(lr_model.coef_))
print("Intercept: {}".format(lr_model.intercept_))

plt.scatter(x_train, y_train, color = 'red', label = 'Training Set')
plt.plot(x_test, lr_model.predict(x_test_), color = 'green', label = 'Prediction')
plt.title('GDP Growth (Annual %)')
plt.xlabel('Year')
plt.ylabel('Percentage Growth')
plt.legend()
plt.grid()
plt.show()
```
