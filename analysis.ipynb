## IMPORTING THE DATASET

```python
df = pd.read_csv('data/gdpdata.csv') # Importing the csv file as a dataframe
```

```python
df['GDP deflator: linked series (base year varies by country)']
```

```python
df['GDP deflator: linked series (base year varies by country)'] = df['GDP deflator: linked series (base year varies by country)'].replace('..', 0)
```

```python
df['GDP deflator: linked series (base year varies by country)']
```

```python
df['GDP deflator: linked series (base year varies by country)'] = df['GDP deflator: linked series (base year varies by country)'].astype(float)
```

```python
plt.plot(df['Year'], df['GDP growth (annual %)'], '--.g', label = 'Annual GDP Growth %')
plt.xlabel('Year')
plt.ylabel('Percentage Growth')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(df['Year'], df['GDP per capita (constant LCU)'], '--.b', label = 'GDP Per Capita INR')
plt.xlabel('Year')
plt.ylabel('GDP per Capita in INR')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(df['Year'], df['GDP per capita (current LCU)'], '--.r', label = 'GDP Per Capita current INR')
plt.xlabel('Year')
plt.ylabel('GDP Per Capita in current INR')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(df['Year'], df['GDP (current LCU)'], '--.g', label = 'GDP current INR')
plt.xlabel('Year')
plt.ylabel('GDP in current INR')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(df['Year'], df['GDP (constant LCU)'], '--.c', label = 'GDP constant INR')
plt.xlabel('Year')
plt.ylabel('GDP in constant INR')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(df['Year'], df['GDP deflator (base year varies by country)'], '--.m', label = 'GDP Deflator')
plt.xlabel('Year')
plt.ylabel('GDP Deflator')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(df['Year'], df['GDP deflator: linked series (base year varies by country)'], '--.y', label = 'GDP deflator Linked series')
plt.xlabel('Year')
plt.ylabel('GDP deflator Linked series')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(df['Year'], df['GDP per capita growth (annual %)'], '--.w', label = 'Annual GDP Per Capita Growth %')
plt.xlabel('Year')
plt.ylabel('Percentage Growth')
plt.legend()
plt.grid()
plt.show()
```
